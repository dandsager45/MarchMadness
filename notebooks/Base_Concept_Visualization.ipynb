```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Simulate data for two features (e.g., points scored vs. points allowed)
np.random.seed(42)
feature_1 = np.random.normal(loc=75, scale=5, size=100)  # Points scored
feature_2 = np.random.normal(loc=70, scale=5, size=100)  # Points allowed
labels = feature_1 - feature_2 > 0  # Winners if points scored > points allowed

# Create a DataFrame
data = pd.DataFrame({
    'Points Scored': feature_1,
    'Points Allowed': feature_2,
    'Winner': labels
})

# Plot using Seaborn
sns.set_theme(style="whitegrid")
plt.figure(figsize=(8, 6))
scatter = sns.scatterplot(
    data=data,
    x='Points Scored',
    y='Points Allowed',
    hue='Winner',
    palette='coolwarm',
    s=100
)

# Add a simple decision boundary (e.g., line where points scored = points allowed)
x_vals = np.linspace(data['Points Scored'].min(), data['Points Scored'].max(), 100)
plt.plot(x_vals, x_vals, '--', color='green', label='Decision Boundary')

# Customize plot
plt.title('Team Performance with Decision Boundary', fontsize=16)
plt.xlabel('Points Scored', fontsize=12)
plt.ylabel('Points Allowed', fontsize=12)
plt.legend(title='Winner', loc='upper left')
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Simulate data for points scored
np.random.seed(42)
points_scored = np.random.normal(loc=75, scale=10, size=10)  # Points scored by 10 teams

# Create a DataFrame
data = pd.DataFrame({
    'Team': [f'Team {i+1}' for i in range(len(points_scored))],
    'Points Scored': points_scored
})

# Calculate the deltas (differences between each pair of teams)
# We will calculate the difference between each team and the first team (Team 1)
deltas = data['Points Scored'] - data['Points Scored'].iloc[0]

# Plotting the deltas
sns.set_theme(style="whitegrid")
plt.figure(figsize=(8, 6))
sns.scatterplot(
    data=data,
    x='Team',
    y=deltas,
    s=100
)

# Customize plot
plt.title('Deltas Between Points Scored (Relative to Team 1)', fontsize=16)
plt.xlabel('Teams', fontsize=12)
plt.ylabel('Points Scored Delta', fontsize=12)
plt.xticks(rotation=45)  # Rotate team names for better readability
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Simulate data for two features: points scored and offensive rebounds
np.random.seed(42)
points_scored = np.random.normal(loc=75, scale=10, size=10)  # Points scored by 10 teams
offensive_rebounds = np.random.normal(loc=10, scale=3, size=10)  # Offensive rebounds by 10 teams

# Create a DataFrame
data = pd.DataFrame({
    'Team': [f'Team {i+1}' for i in range(len(points_scored))],
    'Points Scored': points_scored,
    'Offensive Rebounds': offensive_rebounds
})

# Calculate the deltas (differences between each team and Team 1)
team_1_points = data['Points Scored'].iloc[0]
team_1_rebounds = data['Offensive Rebounds'].iloc[0]

data['Points Scored Delta'] = data['Points Scored'] - team_1_points
data['Offensive Rebounds Delta'] = data['Offensive Rebounds'] - team_1_rebounds

# Create a 3D plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plotting the deltas
ax.scatter(
    data['Points Scored Delta'], 
    data['Offensive Rebounds Delta'], 
    np.zeros_like(data['Points Scored Delta']),  # Z axis for team names or some third dimension (fixed for simplicity)
    s=100
)

# Set labels and title
ax.set_xlabel('Points Scored Delta')
ax.set_ylabel('Offensive Rebounds Delta')
ax.set_zlabel('Teams')  # Z label (can be customized as a third feature)

# Add team names as text annotations
for i in range(len(data)):
    ax.text(
        data['Points Scored Delta'][i],
        data['Offensive Rebounds Delta'][i],
        0,  # Z value for text position
        data['Team'][i],
        size=10,
        color='black'
    )

plt.title('3D Deltas in Points Scored and Offensive Rebounds (Relative to Team 1)', fontsize=16)
plt.show()
```
